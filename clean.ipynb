```python
import pandas as pd
import numpy as np
```

```python
path = './data.csv'
df = pd.read_csv(path)
df.drop_duplicates
df[['song', 'artist']] = df['name'].str.split("*", 1, expand = True)
df.drop(df.columns[0], axis=1, inplace=True)
df_cols = df.columns.tolist()[-2:] + df.columns.tolist()[:-2] 
df = df[df_cols]
df.to_csv('./clean_data.csv')
```
